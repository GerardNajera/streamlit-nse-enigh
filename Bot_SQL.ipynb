```python
import pandas as pd

#Cargamos el archivo

df=pd.read_excel("BD_ENIGH_24.xlsx")
df.head(5)
```

```python
import sqlite3
from sqlalchemy import create_engine

#Creamos un motor de base de datos SQLite con sqlalchemy
engine= create_engine("sqlite:///DB_ENIGH.db")

df.to_sql("ENIGH", engine,if_exists="replace", index=False)
```

```python
#Importar dotenv para cargar las variables de entorno
from dotenv import load_dotenv
import os

from langchain_community.utilities import SQLDatabase
from langchain_community.agent_toolkits import create_sql_agent
from langchain_openai import ChatOpenAI

#cargar variables de entorno desde archivo env
load_dotenv()

#conectar el agente LangChain con la base de datos SQL
db= SQLDatabase(engine=engine)

#Intanciamos el LLM
llm= ChatOpenAI(model="gpt-4o",temperature=0, openai_api_key=os.getenv('OPENAI_API_KEY'))

#Creamos el agente
agent_executor = create_sql_agent (llm, db=db, agent_type="openai-tools", verbose=True)
```

```python
response = agent_executor.invoke ({"input": "¿Cuantos registros hay en la base?"})
print (response["output"])
```
